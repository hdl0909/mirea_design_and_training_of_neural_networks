```python
import pandas as pd
import numpy as np
```

```python
data = {
    'x1': [1, 1, 0, 0],
    'x2': [0, 0, 0, 1],
    'x3': [1, 0, 0, 0],
    'x4': [0, 0, 1, 1]
}
data = pd.DataFrame(data, dtype='float')
data
```

```python
class nnKohonen:

    def __init__(self, learning_rate=0.6, epochs=50):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None

    def fit(self, data):
        # n_features = data.shape[1] # кол-во обучаемых векторов
        
        # np.random.seed(42)
        # self.weights = np.random.rand(2, n_features)

        self.weighted = np.array([[0.2, 0.6, 0.5, 0.9],
                                [0.8, 0.4, 0.7, 0.3]])

        
        nu = self.learning_rate
        for _ in range(self.epochs):
            for train_x in data:
                d1 = np.sqrt(np.sum(np.pow((data[train_x] - self.weighted[0]), 2)))
                d2 = np.sqrt(np.sum(np.pow((data[train_x] - self.weighted[1]), 2)))
                n_winner = np.argmin([d1, d2])
                self.weighted[n_winner] = self.weighted[n_winner] + nu * (data[train_x] - self.weighted[n_winner])
            if nu > 0.05:
                nu -= 0.01

    def predict(self, data):
        clusters = []
        for x in data:
            d1 = np.sqrt(np.sum(np.pow((data[x] - self.weighted[0]), 2)))
            d2 = np.sqrt(np.sum(np.pow((data[x] - self.weighted[1]), 2)))
            n_winner = np.argmin([d1, d2])
            clusters.append(n_winner)
        return np.array(clusters)
```

```python
nn = nnKohonen(learning_rate=0.6, epochs=50)
```

```python
nn.fit(data)
```

```python
np.round(nn.weighted, 2)
```

```python
nn.predict(data)
```
