```python
import pandas as pd
```

```python
train = {'X1': pd.Series([1, 1, -1, -1]),
         'X2': pd.Series([1, -1, 1, -1])}
X = pd.DataFrame(train)
target =  pd.Series([1, -1, -1, -1])
```

```python
X
```

```python
lr = 1
w11 = 0
w21 = 0
T = 0
```

```python
def delta(x1, x2, w11, w21, T, lr, y):
    NET = x1 * w11 + x2 * w21 - T
    if NET > 0:
        y0 = 1
    else:
        y0 = -1
    
    if y0 == y:
        return w11, w21, T
    
    e = y - y0
    w11 = w11 + lr * e * x1
    w21 = w21 + lr * e * x2
    T = T + lr * e * (-1)
    return w11, w21, T
```

```python
for input in range(X.shape[0]):
    w11, w21, T = delta(X.iloc[input, 0], X.iloc[input, 1], w11, w21, T, lr, target[input])
```

```python
def predict(x1, x2, w11, w21):
    return 1 if (x1*w11 + x2*w21 - T) > 0 else -1
```

```python
print(w11, w21, T)
```

```python
for i in range(X.shape[0]):
    print(f'{X.iloc[i, 0]}, {X.iloc[i, 1]}-> {predict(X.iloc[i, 0], X.iloc[i, 1], w11, w21)}') 
```
