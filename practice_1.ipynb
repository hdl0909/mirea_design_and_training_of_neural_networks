```python
import pandas as pd
import numpy as np
```

```python
y = pd.Series([-1, -1 , -1, 1])
X = pd.DataFrame({'X1': [-1, 1, -1, 1], 'X2': [-1, -1, 1, 1]})
```

```python
X
```

```python
w11 = np.random.rand() / 100
w21 = np.random.rand() / 100
T = np.random.rand() / 100
```

```python
def delta(x1, x2, w11, w21, T, y):
    w11 = w11 + x1*y
    w21 = w21 + x2*y
    T = T - y
    return w11, w21, T
```

```python
for i in range(X.shape[0]):
    X12 = X.iloc[i, :].to_list()
    w11, w21, T = delta(X12[0], X12[1], w11, w21, T, y[i])
```

```python
def predict(x1, x2, w11, w21):
    return (x1*w11 + x2*w21 - T) > 0 
```

```python
predict(1, -1, w11, w21)
```

```python
print(w11, w21, T)
```

```python
for i in range(X.shape[0]):
    X12 = X.iloc[i, :].to_list()
    print(f'{X12[0]} {X12[1]}: {predict(X12[0], X12[1], w11, w21)}')
```
